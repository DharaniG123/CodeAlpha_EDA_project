```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Student": ["Asha", "Rahul", "Priya", "Arjun", "Meera"],
    "Marks": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

plt.bar(df["Student"], df["Marks"])
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
